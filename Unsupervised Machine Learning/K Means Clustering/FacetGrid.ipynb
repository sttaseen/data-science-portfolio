**FacetGrid Testing**

This is just a testing ground to test out how facetgrid works and what not.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.pairplot(data=tips, hue = "sex")
```

```python
g = sns.FacetGrid(data=tips, hue='sex', height=6, palette='coolwarm')
g = g.map_dataframe(sns.histplot, x='tip').add_legend()
```

```python
sns.histplot(data = tips, x='tip')
```
